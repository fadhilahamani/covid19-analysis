Pada bagian ini, dataset akan diolah sesuai dengan problem statement yang telah ditentukan.

## Negara yang memiliki jumlah pasien terkonfirmasi COVID-19 paling banyak dan paling sedikit
Untuk mencari data negara yang memiliki jumlah pasien terkonfirmasi COVID-19 paling banyak dan paling sedikit, harus dicari terlebih dahulu total kasus yang terkonfirmasi di setiap negara dengan menggunakan `.groupby('country')`.

Lalu data-data yang ada pada kolom confirmed dapat dijumlahkan dengan `.sum()`

Lalu ubah data-data tersebut menjadi dataframe dengan nama kolom-kolomnya adalah 'country' dan 'total confirmed' dan gunakan `.sort_values('total confirmed'), ascending = False)` untuk mengurutkan data dari yang terbesar ke terkecil.

```python
data_confirmed = data.groupby('country')['confirmed'].sum()
data_confirmed = data_confirmed.to_frame().reset_index()
data_confirmed.columns = ['country', 'total confirmed']

data_confirmed.sort_values('total confirmed', ascending = False)
```

Berikut ini visualisasi untuk data total kasus covid-19 yang terkonfirmasi di setiap negara

```python
data_confirmed.plot(x = 'country',
                    y = 'total confirmed',
                    title = 'Total Kasus COVID-19 yang Terkonfirmasi',
                    figsize = (15,5),
                    kind = 'bar')
```

Negara paling banyak kasus terkonfirmasi COVID-19 adalah India, perkembangan kasusnya adalah sebagai berikut

```python
data[data ['country'] == 'India']
```

## 10 Kota (province state) yang memiliki jumlah kasus kematian tertinggi
Untuk mencari data 10 kota yang memiliki jumlah kasus kematian tertinggi, harus dicari terlebih dahulu total kasus kematian di setiap kota dengan menggunakan `.groupby('province state')`.

Lalu data-data yang ada pada kolom deaths dapat dijumlahkan dengan `.sum()`

Lalu ubah data-data tersebut menjadi dataframe dengan nama kolom-kolomnya adalah 'country' dan 'total death' dan gunakan `.sort_values('total death'), ascending = False).head(10)` untuk mengurutkan data dari yang terbesar ke terkecil dan menampilkan 10 data terbesar.


```python
data_death = data.groupby('province state')['deaths'].sum()
data_death = data_death.to_frame().reset_index()
data_death.columns = ['province state', 'total death']

data_death.sort_values('total death', ascending = False).head(10)
```

Berikut ini visualisasi untuk data 10 kota dengan kasus kematian paling banyak

```python
death_10 = data_death.sort_values('total death', ascending = False).head(10)
death_10.plot(x = 'province state',
              y = 'total death',
              title = '10 Kota dengan Kasus Kematian Covid-19 paling banyak',
              figsize = (15,5),
              kind = 'scatter')
```

## Rata-rata jumlah kasus aktif Covid-19 pada setiap negara
Untuk mencari data nata-rata jumlah kasus aktif covid-19 pada setiap negara, harus dikelompokan terlebih dengan menggunakan `.groupby('country')`.
Lalu tambahkan `.mean()` untuk menghitung rata-rata.



```python
data_active = data.groupby('country')['active'].mean()
data_active = data_active.to_frame().reset_index()
data_active.columns = ['country', 'rata-rata kasus aktif']

data_active.sort_values('rata-rata kasus aktif', ascending = False)
```

Berikut ini visualisasi untuk data rata-rata kasus aktif covid-19 di setiap negara

```python
data_active.plot(x = 'country',
                 y = 'rata-rata kasus aktif',
                 title = 'Rata-rata kasus aktif Covid-19',
                 figsize = (15,5),
                 kind = 'line')
```
